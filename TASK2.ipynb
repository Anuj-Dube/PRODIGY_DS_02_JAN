```python
# Libaries 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Loading dataset 
data = pd.read_csv('CancerSurvival.csv')
data.head()
```

```python
#Checking the dataset
data.info()
```

```python
data.describe()
```

```python
#Step 4:- Perform Data Cleaning
print(data.isnull().sum())

# Fill missing values if there any 
data['year'].fillna(data['year'].median(), inplace=True)
```

```python
# Step 5: Exploratory Data Analysis (EDA)
data.hist(bins=10, figsize=(10, 6))
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='survival_rate', data=data)
plt.title('Distribution of Survival')
plt.show()
```

```python
counts = data.groupby(['cancer_type', 'race']).size().unstack()
counts.plot(kind="bar")
plt.show()
```

```python
counts = data.groupby(['cancer_type', 'year']).size().unstack()
counts.plot(kind="bar")
plt.show()
```
